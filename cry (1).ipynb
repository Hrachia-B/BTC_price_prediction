```python
# Import libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



# Clean data and convert to numeric
data_cleaned = data.apply(pd.to_numeric, errors='coerce').dropna()

# Add Moving Averages (MA)
windows = [7, 14, 30]  # 7-day, 14-day, and 30-day moving averages
for window in windows:
    data_cleaned[f'MA_{window}'] = data_cleaned['Close'].rolling(window=window).mean()

# Add Volatility (High - Low) over the same windows
for window in windows:
    data_cleaned[f'Volatility_{window}'] = (
        data_cleaned['High'].rolling(window=window).max() -
        data_cleaned['Low'].rolling(window=window).min()
    )

# Add Lag Features
lags = [1, 3, 7]  # 1-day, 3-day, and 7-day lags
for lag in lags:
    data_cleaned[f'Close_Lag_{lag}'] = data_cleaned['Close'].shift(lag)

# Drop rows with NaN values created by rolling and shifting operations
data_cleaned.dropna(inplace=True)

# Plot Price Trends
plt.figure(figsize=(14, 8))
for column in ['Adj Close', 'Close', 'High', 'Low', 'Open']:
    plt.plot(data_cleaned.index, data_cleaned[column], label=column)
plt.title('Price Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Plot Volume Trends
plt.figure(figsize=(14, 6))
plt.plot(data_cleaned.index, data_cleaned['Volume'], color='orange', label='Volume')
plt.title('Volume Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()

# Compute and display Correlation Matrix
correlation_matrix = data_cleaned.corr()
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Cryptocurrency Data with Engineered Features")
plt.show()

# Summary of the feature engineering process
print("Feature Engineering Summary:")
print("Added Moving Averages (7, 14, 30 days).")
print("Added Volatility features (7, 14, 30 days).")
print("Added Lag features (1, 3, 7 days).")
```

```python
from sklearn.preprocessing import MinMaxScaler
# Step 1: Feature Selection
# Dropping features with high correlation (> 0.95) and keeping the most relevant ones
correlation_matrix = data_cleaned.corr()
threshold = 0.95

# Identify features to drop based on the correlation threshold
to_drop = set()
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            to_drop.add(correlation_matrix.columns[i])

# Drop the identified features
selected_features = data_cleaned.drop(columns=list(to_drop))

# Step 2: Normalization using MinMaxScaler
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(selected_features)
normalized_data = pd.DataFrame(normalized_data, columns=selected_features.columns, index=selected_features.index)

# Plot Price Trends
plt.figure(figsize=(14, 8))
for column in ['Adj Close', 'Close', 'High', 'Low', 'Open']:
    if column in normalized_data.columns:
        plt.plot(normalized_data.index, normalized_data[column], label=column)
plt.title('Normalized Price Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Normalized Price')
plt.legend()
plt.show()

# Plot Volume Trends
if 'Volume' in normalized_data.columns:
    plt.figure(figsize=(14, 6))
    plt.plot(normalized_data.index, normalized_data['Volume'], color='orange', label='Volume')
    plt.title('Normalized Volume Trends Over Time')
    plt.xlabel('Date')
    plt.ylabel('Normalized Volume')
    plt.legend()
    plt.show()

# Compute and display Correlation Matrix of Selected Features
correlation_matrix_selected = normalized_data.corr()
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix_selected, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Selected and Normalized Features")
plt.show()

# Summary of the feature engineering and selection process
print("Feature Engineering and Selection Summary:")
print(f"Original features: {len(data_cleaned.columns)}")
print(f"Selected features: {len(selected_features.columns)}")
print("Normalization applied using MinMaxScaler.")
```

```python
from sklearn.model_selection import train_test_split
# Step 3: Data Preparation
# Define target and features
target_index = 0  # Assuming 'Close' is at index 1

X = normalized_data.drop(columns=[normalized_data.columns[target_index]])
y = normalized_data.iloc[:, target_index]

# Create sequences for training, validation, and testing
sequence_length = 30  # Example: 30 days of historical data per sequence

def create_sequences(data, target, seq_length):
    X_seq, y_seq = [], []
    for i in range(len(data) - seq_length):
        X_seq.append(data.iloc[i:i + seq_length].to_numpy())
        y_seq.append(target.iloc[i + seq_length])
    return np.array(X_seq), np.array(y_seq)

# Split data into sequences
X_seq, y_seq = create_sequences(X, y, sequence_length)

# Further split sequences into train, validation, and test sets
train_size = int(len(X_seq) * 0.7)
val_size = int(len(X_seq) * 0.15)

def split_sequences(X_seq, y_seq, train_size, val_size):
    X_train = X_seq[:train_size]
    y_train = y_seq[:train_size]
    X_val = X_seq[train_size:train_size + val_size]
    y_val = y_seq[train_size:train_size + val_size]
    X_test = X_seq[train_size + val_size:]
    y_test = y_seq[train_size + val_size:]
    return X_train, y_train, X_val, y_val, X_test, y_test

X_train, y_train, X_val, y_val, X_test, y_test = split_sequences(X_seq, y_seq, train_size, val_size)
```

```python
from reservoirpy.nodes import Reservoir, Ridge
from sklearn.metrics import mean_squared_error, r2_score
# Ensure target dimensions are (samples, 1)
y_train = y_train.reshape(-1, 1)
y_val = y_val.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

# Reservoir Computing Parameters (adapted for our data)
reservoir_size = 1000  # Size of the reservoir, adjusted to fit our data
leak_rate = 0.25       # Leakage rate (based on original parameters)
density = 0.1          # Connectivity density of the reservoir

# Step 4: Define Reservoir Computing Model
reservoir = Reservoir(units=reservoir_size, lr=leak_rate, rc_connectivity=density)  # Explicitly pass 'units' parameter
readout = Ridge(ridge=1e-6)  # Corrected parameter for regularization

# Manual processing of each sequence
print("Processing sequences through the reservoir...")
reservoir_states_train = []
for sequence in X_train:
    state = reservoir.run(sequence)  # Process each timestep in the sequence
    reservoir_states_train.append(state[-1])  # Use the last state of the sequence
reservoir_states_train = np.array(reservoir_states_train)

reservoir_states_val = []
for sequence in X_val:
    state = reservoir.run(sequence)
    reservoir_states_val.append(state[-1])
reservoir_states_val = np.array(reservoir_states_val)

reservoir_states_test = []
for sequence in X_test:
    state = reservoir.run(sequence)
    reservoir_states_test.append(state[-1])
reservoir_states_test = np.array(reservoir_states_test)

# Verify shapes before training
print(f"Shape of reservoir_states_train: {reservoir_states_train.shape}")
print(f"Shape of y_train: {y_train.shape}")

# Train the readout
print("Training the readout layer...")
readout = readout.fit(reservoir_states_train, y_train)

# Evaluate the model manually
train_predictions = readout.run(reservoir_states_train)
val_predictions = readout.run(reservoir_states_val)

def evaluate_model(y_true, y_pred, dataset_name):
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"{dataset_name} - MSE: {mse:.4f}, R²: {r2:.4f}")
    return mse, r2

# Training evaluation
evaluate_model(y_train, train_predictions, "Training")
# Validation evaluation
evaluate_model(y_val, val_predictions, "Validation")

# Test evaluation
test_predictions = readout.run(reservoir_states_test)
evaluate_model(y_test, test_predictions, "Test")
```

```python
from itertools import product
# Ensure target dimensions are (samples, 1)
y_train = y_train.reshape(-1, 1)
y_val = y_val.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

# Reservoir Computing Parameters for Grid Search
reservoir_sizes = [300, 500, 700, 1000]
leak_rates = [0.1, 0.2, 0.3, 0.5]
ridge_values = [1e-4, 1e-3, 1e-2]
densities = [0.05, 0.1, 0.2]

# Grid Search
best_params = None
best_val_mse = float('inf')

for reservoir_size, leak_rate, ridge, density in product(reservoir_sizes, leak_rates, ridge_values, densities):
    print(f"Testing configuration: Reservoir Size={reservoir_size}, Leak Rate={leak_rate}, Ridge={ridge}, Density={density}")

    # Define Reservoir and Ridge models
    reservoir = Reservoir(units=reservoir_size, lr=leak_rate, rc_connectivity=density)
    readout = Ridge(ridge=ridge)

    # Process training and validation sequences
    def process_sequences(reservoir, X_sequences):
        reservoir_states = []
        for sequence in X_sequences:
            for step in sequence:
                state = reservoir.run(step.reshape(1, -1))  # Process each timestep
            reservoir_states.append(state.flatten())  # Store the final state of the sequence
        return np.array(reservoir_states)

    try:
        reservoir_states_train = process_sequences(reservoir, X_train)
        reservoir_states_val = process_sequences(reservoir, X_val)

        # Train the readout
        readout = readout.fit(reservoir_states_train, y_train)

        # Evaluate on validation set
        val_predictions = readout.run(reservoir_states_val)
        val_mse = mean_squared_error(y_val, val_predictions)

        print(f"Validation MSE: {val_mse:.4f}")

        # Update best parameters if current configuration is better
        if val_mse < best_val_mse:
            best_val_mse = val_mse
            best_params = (reservoir_size, leak_rate, ridge, density)

    except Exception as e:
        print(f"Error with configuration: {e}")

print(f"Best Validation MSE: {best_val_mse:.4f}")
print(f"Best Parameters: Reservoir Size={best_params[0]}, Leak Rate={best_params[1]}, Ridge={best_params[2]}, Density={best_params[3]}")
```
